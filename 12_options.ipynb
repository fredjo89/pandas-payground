### Changing pandas options with Attributes and dot syntax (`pd.options.`)

```python
# Generating a 1000x50 matrix of random integers between 0 and 99
df = pd.DataFrame(np.random.randint(0, 100, [1000, 50]))
df.head(3)
```

```python
pd.options.display.max_rows
pd.options.display.max_columns
```

```python
pd.options.display.max_rows = 5
pd.options.display.max_columns = 4
df.head(3)
```

```python
pd.options.display.max_rows = 5
pd.options.display.max_columns = 400
df.head(500)
```

### Changing pandas Options with Methods

```python
pd.get_option("display.max_rows")
```

```python
pd.set_option("display.max_colwidth", 100)
pd.options.display.max_colwidth = 100
pd.set_option("display.max_columns", 100)
```

### The `precision` Option

```python
df = pd.DataFrame(np.random.randn(5, 5))
display(df)
pd.options.display.precision = 2
display(df)
```
