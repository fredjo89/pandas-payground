### Create a Multiindex with the `set index` Method

```python
bigmac = bigmac.set_index(keys=["Date", "Country"])
bigmac = bigmac.sort_index()
bigmac.head(3)
```

### Extract Index Level Values with the `get_level_values` Method

```python
bigmac = pd.read_csv(
    file_path + "bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"]
).sort_index()
bigmac.head(3)


bigmac.index.get_level_values(level="Date")
bigmac.index.get_level_values(level=0)

bigmac.index.get_level_values(level="Country")
bigmac.index.get_level_values(level=1)
```

### Change Index Level name with the `set_names` Method

```python
bigmac.index.set_names(names=["Day", "Location"])
bigmac.index.set_names(level=0, names="Day")
bigmac.index.set_names(level="Date", names="Day")
```

### The `sort_index` method on a `MultiIndex` DataFrame

```python
bigmac.sort_index(level=["Country", "Date"], ascending=[False, True])
```

### Extract Rows from a `MultiIndex` DataFrame

```python
bigmac.loc[("2010-01-01", "Argentina")]
bigmac.loc[("2010-01-01", "Argentina"), ["Price in US Dollars"]]
bigmac.loc[("2010-01-01", "Argentina"), ["Price in US Dollars", "Price in US Dollars"]]
bigmac.loc[
    [("2010-01-01", "Argentina")], ["Price in US Dollars", "Price in US Dollars"]
]
```

Note that `MultiIndex` does not affect `iloc` at all

### The `transpose` method

It is possible to have column multiIndex as well as row multiIndex

```python
bigmac_t = bigmac.transpose()

bigmac_t.loc[[("Price in US Dollars")], ("2010-01-01",)]
bigmac_t.loc[[("Price in US Dollars")], [("2010-01-01", "Argentina")]]

bigmac_t.loc[
    [("Price in US Dollars")], ("2010-01-01", "Sri Lanka"):("2010-01-01", "Ukraine")
]
```

### The `unstack` Method

`unstack` does the opposite of `stack` <br>
`unstack` is allso called `pivot`

```python
world.head()
```

```python
world.stack().unstack().head()
```

```python
world.unstack().unstack()
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack(0).head(3)
```

```python
s.unstack(-1).head(2)
```

```python
s.unstack("year").head(2)
```

```python
s.unstack([1, 0], fill_value=0).head(3)
```

### The `pivot` Method

```python
sales = pd.read_csv(file_path + "salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")

sales.head(3)
```

```python
sales.pivot(index="Date", columns="Salesman", values="Revenue").head(3)
```

### The `pivot_table()` Method

The `pivot_table()` Method works as the pivot method in Excel

```python
foods.head()
```

```python
foods = pd.read_csv(file_path + "foods.csv")

foods.pivot_table(values="Spend", index="Gender", aggfunc="mean")
```

Doing the same thing as above with `.groupby()`:

```python
foods.groupby("Gender")[["Spend"]].mean()
```

```python
foods.pivot_table(
    values="Spend", index=["Gender", "Item"], columns=["City"], aggfunc=["mean", "sum"]
).head(3)
```
